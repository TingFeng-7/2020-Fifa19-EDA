```python
defense_feature_columns=['Name',
                         'Age',
                         'Position',
                         'Overall',
                         'Reactions',
                         'Interceptions',
                         'Composure',
                         'Marking',
                         'StandingTackle',
                         'SlidingTackle' ,
                         'BallControl',
                         'ShortPassing',
                         'HeadingAccuracy',
                         'Crossing',
                         ]
def detail_position_table(df,feature_columns):
    feature_df=pd.DataFrame(df,columns=feature_columns)
    return feature_df
```

```python
def gen_person(df,Weizhi,label):
    wz_df=df.loc[df['Position']==Weizhi]
    wz_store=wz_df.corr()#存储偏差
    relation=wz_store.loc[wz_store[label]>=0.75]##label：Overall 留作扩展
    useful_columns=relation.index.tolist()
    print(Weizhi,'Feature columns:',useful_columns,'\n')
    return wz_df,useful_columns ###可以用这个df去画scatter
```

```python
lb_df,lb_key_columns=gen_person(de,'LB','Overall')
#px.scatter_matrix(lb_df,dimensions=lb_key_columns)
```
